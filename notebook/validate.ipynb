```python
import pandas as pd
```

```python
df = pd.read_csv("/config/workspace/artifact/12272022__142444/data_ingestion/dataset/train.csv")
```

```python
# gives the percentage of missing values in each columns along with column name
df.isnull().sum()/df.shape[0]
```

```python
# gives the values of missing value percentage
(df.isnull().sum()/df.shape[0]).values
```

```python
# length of the values
(df.isnull().sum()/df.shape[0]).values.__len__()
```

```python
# to find the column names
(df.isnull().sum()/df.shape[0]).index
```

```python
# to print the column name and missing percentage
for column_name, missing_percentage in zip((df.isnull().sum()/df.shape[0]).index , (df.isnull().sum()/df.shape[0]).values):
    print(column_name,missing_percentage * 100)
```

```python
# exmaale for zip function
l1 = [1,2,3,4]
l2 = [5,6,7,8]

for l1_val, l2_val in zip(l1,l2):
    print(l1_val, l2_val)
```

```python
# to find the list of column names which need to be dropped
drop_column_names = []
for column_name, missing_percentage in zip((df.isnull().sum()/df.shape[0]).index , (df.isnull().sum()/df.shape[0]).values):
    print(column_name, missing_percentage)
    if missing_percentage > 0.3:
        drop_column_names.append(column_name)
```
